```python
# 리워드 상세정보, 키워드 추출
kkma = Kkma()
# 끊긴 위치를 저장하기 위함. unknown 시작하는 위치.txt , 내용: 이미 한 url들
offset = 6000 # 다시 시작할 위치

# 다들 아시겠지만 그냥 써놉니다! range안에 수정해서 본인이 맡은 번호 ㄱㄱ하기

# len(new_links)
with open("infos.txt", "a") as f:
  for idx in range(0, 20): # range() -> 이거 수정하기
      print(idx)
      url = new_links[idx]
      resp = requests.get(url)
      html = resp.text
      soup = BeautifulSoup(html, 'html.parser')
      # driver.get(url)

      time.sleep(1)
      url = new_links[idx]
      code = url.split('/')[-1]
      # 반환정책 url
      url = f'https://www.wadiz.kr/web/campaign/detail/fundingInfo/{code}?_refer_section_st=REWARD_1' # 반환정책 url
      driver.get(url)
      # 리워드 상세정보
      sub_category = driver.find_element(By.XPATH, '//*[@id="reward-static-product-content-app"]/div/div/p/span').text
      # 스토리 url
      url = f'https://www.wadiz.kr/web/campaign/detail/{code}?_refer_section_st=REWARD_1'
      driver.get(url)
      # 제목
      title = driver.find_element(By.XPATH, '//*[@id="container"]/div[3]/h2/a').text
      # title = re.sub('[^가-힣]', ' ', title)
      # 본문 내용
      explain = driver.find_element(By.XPATH, '//*[@id="container"]/div[6]/div/div[1]/div[2]/div/div/section/div[3]').text
      # explain = re.sub('[^가-힣]', ' ', explain)
      # 공통키워드
      
      a = kkma.nouns(title)
      b = kkma.nouns(explain)
      key_word = []
      for i in a:
        if i in b:
          key_word.append(i)
      key_word = ' '.join(key_word)
      print(f'code : {code} / sub_category: {sub_category} / key_word : {key_word}')
      print(f'{code};{sub_category};{key_word}', file=f)
```
