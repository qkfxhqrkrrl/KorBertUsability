```python
for row in range(len(df)):
    df.iloc[row, 9] = int(df.iloc[row, 9].replace(",", ""))
df.loc[:, "cnt_supporters"].astype(int)
```

```python
criterion = {
    "0" : "0",
    "500" : "1",
    "1000" : "2",
    "1500" : "3",
    "2000" : "4",
    "2500" : "5",
    "3000" : "6",
}

for crit, rank in criterion.items():
    df.loc[df.loc[:, "cnt_supporters"] > int(crit) ,"rank_cnt_supporters"] = rank
```

```python
df.loc[:, ["cnt_supporters", "rank_cnt_supporters"]]
```

```python
suc_by_cate2 = df.groupby(["rank_cnt_supporters",'success']).size().unstack()
suc_by_cate2.fillna(0, inplace=True)
# suc_by_cate2.loc[:,"sum"] = suc_by_cate2.iloc[:, 0] +suc_by_cate2.loc[:, 1]
# # print(suc_by_cate2.columns) 
# suc_by_cate2.sort_values("sum", inplace=True, ascending=False)
# del suc_by_cate2["sum"]
suc_by_cate2.plot.bar(stacked=True, figsize=(16,9))
plt.xticks(rotation=0, fontsize=16, ticks =[0,1,2,3,4,5,6] ,labels=["<500", "500~1000", "1000~1500", "1500~2000", "2000~2500", "2500~3000", ">3000"])
plt.xlabel("서포터 수에 대한 프로젝트")
```

```python
df["success_rate"].value_counts
```

```python
criterion = {
    "0" : "0",
    "100" : "1",
    "500" : "2",
    "1000" : "3",
    "5000" : "4",
    "10000" : "5"
}

for crit, rank in criterion.items():
    df.loc[df.loc[:, "success_rate"] > int(crit) ,"rank_success_rate"] = rank
```

```python
df.rank_success_rate.value_counts()
```

```python
sub_df = df.groupby(["rank_success_rate",'category']).size().unstack().fillna(0)
# fig = plt.figure(figsize=[16, 9*len(df.category.unique())])

for idx, cate in enumerate(df.category.unique()):
    # ax = fig.add_subplot(len(df.category.unique()), 1, idx+1)
    ratio = []
    cate_sum = sub_df[cate].sum()
    # print(cate, " : ", cate_sum)
    for value in sub_df[cate]:
        ratio.append(value/cate_sum)
    # print(ratio)
    crit = {
      0 : "<100%", 
      1 : "100%~500%",
      2 : "500%~1000%",
      3 : "1000%~5000%",
      4 : "5000%~10000%",
      5 : ">10000%",
    }
    # labels = [crit[rank] for rank in sub_df[cate].index]
    labels = [crit[int(rank)] for rank in sub_df[cate].index]
    # ax.pie(ratio, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)
    # ax.set_title(cate+"({})".format(int(cate_sum)))
    plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)
    plt.title(cate+"({})".format(int(cate_sum)))
    plt.savefig("Test{}".format(idx))
    plt.show()
```
